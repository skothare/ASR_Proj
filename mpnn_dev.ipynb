## Task & Description
HIV
Dataset Description: The HIV dataset was introduced by the Drug Therapeutics Program (DTP) AIDS Antiviral Screen, which tested the ability to inhibit HIV replication for over 40,000 compounds. From MoleculeNet.

Task Description: Binary classification. Given a drug SMILES string, predict its activity against HIV virus.

Dataset Statistics: 41,127 drugs.

Dataset Split: Random Split Scaffold Split
- from https://tdcommons.ai/single_pred_tasks/hts/?utm_source#hiv:~:text=BY%204.0.-,HIV,benchmark%20for%20molecular%20machine%20learning.%E2%80%9D%20Chemical%20science%209.2%20(2018)%3A%20513%2D530.,-Dataset%20License%3A

Caco-2 (Cell Effective Permeability), Wang et al.
Dataset Description: The human colon epithelial cancer cell line, Caco-2, is used as an in vitro model to simulate the human intestinal tissue. The experimental result on the rate of drug passing through the Caco-2 cells can approximate the rate at which the drug permeates through the human intestinal tissue.

Task Description: Regression. Given a drug SMILES string, predict the Caco-2 cell effective permeability.

Dataset Statistics: 906 drugs.

Dataset Split: Random Split Scaffold Split
- from https://tdcommons.ai/single_pred_tasks/adme/#caco-2-cell-effective-permeability-wang-et-al:~:text=site%20of%20action.-,Caco%2D2%20(Cell%20Effective%20Permeability)%2C%20Wang%20et%20al.,Journal%20of%20Chemical%20Information%20and%20Modeling%202016%2056%20(4)%2C%20763%2D773,-Dataset%20License%3A

```python
# Download the data for classification (HTS) and regression (ADME)
HIV_data = HTS(name = 'HIV') #https://tdcommons.ai/single_pred_tasks/hts/?utm_source#hiv:~:text=BY%204.0.-,HIV,benchmark%20for%20molecular%20machine%20learning.%E2%80%9D%20Chemical%20science%209.2%20(2018)%3A%20513%2D530.,-Dataset%20License%3A

HIV_df = HIV_data.get_data()

print(f"Shape of HIV data: {HIV_df.shape}")

display(HIV_df.head())
# Y in HIV (classification is a regression label: )
```

```python
# Identify the files and label the columsn
def infer_columns(df):
    smiles_col = None
    label_col = None

    for c in df.columns:
        if c.lower() in ["drug", "smiles", "molecule"]:
            smiles_col = c
        if c.lower() in ["y", "label", "target"]:
            label_col = c

    return smiles_col, label_col

hiv_smiles_col, hiv_label_col = infer_columns(HIV_df)

print("HIV:", hiv_smiles_col, hiv_label_col)
```

```python
# Check SMILES with RDKit
def parse_smiles(smiles):
    if pd.isna(smiles):
        return None
    try:
        return Chem.MolFromSmiles(smiles)
    except Exception:
        return None

hiv_mols = HIV_df[hiv_smiles_col].apply(parse_smiles)

print("Valid HIV molecules:", hiv_mols.notna().sum(), "/", len(hiv_mols))

hiv_invalid = HIV_df.loc[hiv_mols.isna(), [hiv_smiles_col, hiv_label_col]].head()

display(hiv_invalid)
```

## Construct Graphs for GNN from SMILES

```python
from chemprop import data, featurizers
from chemprop.models import MPNN
from chemprop.nn import AtomMessagePassing, MeanAggregation, BinaryClassificationFFN

# ── Load data ──────────────────────────────────────────────────────────
smiles_data = [data.MoleculeDatapoint.from_smi(smi, y=[label])
               for smi, label in zip(HIV_df["Drug"], HIV_df["Y"])]

# ── Train/val/test split ───────────────────────────────────────────────
n         = len(smiles_data)
train_end = int(0.8 * n)
val_end   = int(0.9 * n)

train_data = data.MoleculeDataset(smiles_data[:train_end])
val_data   = data.MoleculeDataset(smiles_data[train_end:val_end])
test_data  = data.MoleculeDataset(smiles_data[val_end:])

# ── Featurizer ─────────────────────────────────────────────────────────
featurizer = featurizers.SimpleMoleculeMolGraphFeaturizer()
train_data.featurizer = featurizer
val_data.featurizer   = featurizer
test_data.featurizer  = featurizer

train_loader = data.build_dataloader(train_data, shuffle=True)
val_loader   = data.build_dataloader(val_data,   shuffle=False)
test_loader  = data.build_dataloader(test_data,  shuffle=False)

# ── Class imbalance weight ─────────────────────────────────────────────
labels     = [smiles_data[i].y[0] for i in range(train_end)]
num_pos    = int(sum(labels))
num_neg    = len(labels) - num_pos
#pos_weight = torch.tensor([min(num_neg / num_pos, 5.0)])   # cap at 5 instead of 29
#print(f"pos_weight: {pos_weight.item():.2f}")

from chemprop.nn import AtomMessagePassing, MeanAggregation, BinaryClassificationFFN, BinaryAUPRC, BinaryAUROC
# ── Model with both AUROC and AUPRC metrics ────────────────────────────
from chemprop.nn import BondMessagePassing

agg = MeanAggregation()
mp = BondMessagePassing(
    dropout=0.2,        # ← add dropout inside message passing
)

ffn = BinaryClassificationFFN(
    dropout=0.3,        # ← more dropout in FFN
)

model = MPNN(
    message_passing=mp,
    agg=agg,
    predictor=ffn,
    metrics=[BinaryAUROC(), BinaryAUPRC()],
    init_lr=1e-4,
    max_lr=5e-4,
    final_lr=1e-4,
    warmup_epochs=5,
)
#model.criterion.pos_weight = pos_weight

# ── Callbacks ──────────────────────────────────────────────────────────
early_stop = pl.callbacks.EarlyStopping(
    monitor="val_loss", patience=10, mode="min"
)
checkpoint = pl.callbacks.ModelCheckpoint(
    monitor="val_loss", save_top_k=1, mode="min", filename="best"
)
```

## Train GNN on Binary Classification Task for HIV

```python
from lightning.pytorch.loggers import WandbLogger

wandb_logger = WandbLogger(
    project="CMU_Automation_S26",   # project name on wandb
    name="chemprop-dmpnn-2",        # this run's name
    log_model=True                # saves model checkpoints to wandb
)

trainer = pl.Trainer(
    max_epochs=200,
    accelerator="auto",
    callbacks=[early_stop, checkpoint],
    enable_progress_bar=True,
    gradient_clip_val=0.5,
    logger=wandb_logger,
)

trainer.fit(model, train_loader, val_loader)

wandb.finish()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (confusion_matrix, roc_auc_score, roc_curve,
                             precision_recall_curve, average_precision_score,
                             ConfusionMatrixDisplay, f1_score)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ── Load best checkpoint ───────────────────────────────────────────────
model = MPNN.load_from_checkpoint(checkpoint.best_model_path)
model = model.to(device)
model.eval()
print(f"Loaded best checkpoint: {checkpoint.best_model_path}")

# ── Init wandb for evaluation ──────────────────────────────────────────
wandb.init(
    project="CMU_Automation_S26",
    name="chemprop-dmpnn-eval",
    job_type="evaluation"
)

# ── Tune threshold on val set ──────────────────────────────────────────
val_probs, val_labels = [], []
with torch.no_grad():
    for batch in val_loader:
        bmg            = batch.bmg
        bmg.V          = bmg.V.to(device)
        bmg.E          = bmg.E.to(device)
        bmg.edge_index = bmg.edge_index.to(device)
        bmg.batch      = bmg.batch.to(device)
        preds = model(bmg, None, None)
        probs = torch.sigmoid(preds).cpu().numpy().flatten()
        val_probs.extend(probs)
        val_labels.extend(batch.Y.cpu().numpy().flatten())

# ── Recall-targeted threshold selection ───────────────────────────────
val_probs_arr  = np.array(val_probs)
val_labels_arr = np.array(val_labels)

precisions, recalls, thresholds_prc = precision_recall_curve(val_labels_arr, val_probs_arr)

target_recall = 0.80  # ← adjust this based on your tolerance

valid = [
    (p, r, t)
    for p, r, t in zip(precisions, recalls, thresholds_prc)
    if r >= target_recall
]

if valid:
    best           = max(valid, key=lambda x: x[0])  # highest precision at target recall
    best_threshold = best[2]
    print(f"Threshold: {best_threshold:.3f} → Recall={best[1]:.2f}, Precision={best[0]:.2f}")
else:
    best_idx       = np.argmax(recalls[:-1])
    best_threshold = float(thresholds_prc[best_idx])
    print(f"Could not achieve recall={target_recall} — using max recall threshold")
    print(f"Threshold: {best_threshold:.3f} → Recall={recalls[best_idx]:.2f}, Precision={precisions[best_idx]:.2f}")

# ── Test predictions ───────────────────────────────────────────────────
all_probs, all_labels = [], []
with torch.no_grad():
    for batch in test_loader:
        bmg            = batch.bmg
        bmg.V          = bmg.V.to(device)
        bmg.E          = bmg.E.to(device)
        bmg.edge_index = bmg.edge_index.to(device)
        bmg.batch      = bmg.batch.to(device)
        preds = model(bmg, None, None)
        probs = torch.sigmoid(preds).cpu().numpy().flatten()
        all_probs.extend(probs)
        all_labels.extend(batch.Y.cpu().numpy().flatten())

all_preds = (np.array(all_probs) > best_threshold).astype(int)

# ── Compute metrics ────────────────────────────────────────────────────
auroc    = roc_auc_score(all_labels, all_probs)
auprc    = average_precision_score(all_labels, all_probs)
baseline = num_pos / (num_pos + num_neg)

# actual recall and precision at chosen threshold
from sklearn.metrics import recall_score, precision_score
actual_recall    = recall_score(all_labels, all_preds)
actual_precision = precision_score(all_labels, all_preds)
actual_f1        = f1_score(all_labels, all_preds)

print(f"\nTest Results:")
print(f"AUROC     : {auroc:.4f}")
print(f"AUPRC     : {auprc:.4f}  (random baseline: {baseline:.4f})")
print(f"Recall    : {actual_recall:.4f}")
print(f"Precision : {actual_precision:.4f}")
print(f"F1        : {actual_f1:.4f}")
print(f"Threshold : {best_threshold:.4f}")

# ── Local plots ────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

cm = confusion_matrix(all_labels, all_preds)
ConfusionMatrixDisplay(cm).plot(ax=axes[0])
axes[0].set_title("Confusion Matrix")

fpr, tpr, _ = roc_curve(all_labels, all_probs)
axes[1].plot(fpr, tpr, label=f"AUC = {auroc:.3f}", color="steelblue")
axes[1].plot([0, 1], [0, 1], 'k--', label="Random")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].set_title("ROC Curve")
axes[1].legend()

precision_curve, recall_curve, _ = precision_recall_curve(all_labels, all_probs)
axes[2].plot(recall_curve, precision_curve, label=f"AUPRC = {auprc:.3f}", color="darkorange")
axes[2].axhline(y=baseline, color='k', linestyle='--', label=f"Random = {baseline:.3f}")
axes[2].axvline(x=actual_recall, color='green', linestyle='--', label=f"Recall = {actual_recall:.2f}")
axes[2].scatter([actual_recall], [actual_precision], color='green', zorder=5)
axes[2].set_xlabel("Recall")
axes[2].set_ylabel("Precision")
axes[2].set_title("Precision-Recall Curve")
axes[2].legend()

plt.tight_layout()
plt.show()

# ── Log to wandb ───────────────────────────────────────────────────────
wandb.log({
    "test/auroc":          auroc,
    "test/auprc":          auprc,
    "test/recall":         actual_recall,
    "test/precision":      actual_precision,
    "test/f1":             actual_f1,
    "test/best_threshold": float(best_threshold),
    "test/target_recall":  target_recall,
    "baseline/auprc":      baseline,

    "test/confusion_matrix": wandb.plot.confusion_matrix(
        y_true=[int(l) for l in all_labels],
        preds=[int(p) for p in all_preds],
        class_names=["inactive", "active"]
    ),
    "test/roc_curve": wandb.plot.roc_curve(
        y_true=[int(l) for l in all_labels],
        y_probas=np.column_stack([1 - np.array(all_probs), np.array(all_probs)]),
        labels=["inactive", "active"]
    ),
    "test/pr_curve": wandb.plot.pr_curve(
        y_true=[int(l) for l in all_labels],
        y_probas=np.column_stack([1 - np.array(all_probs), np.array(all_probs)]),
        labels=["inactive", "active"]
    ),
})

wandb.finish()
```
